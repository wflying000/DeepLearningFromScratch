```python
import json
import matplotlib.pyplot as plt
```

```python
iter_list = json.load(open("../outputs/iter_list.json"))
train_acc_list = json.load(open("../outputs/train_acc_list.json"))
train_loss_list = json.load(open("../outputs/train_loss_list.json"))
test_acc_list = json.load(open("../outputs/test_acc_list.json"))
```

```python
len(train_loss_list)
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.plot(range(len(train_loss_list)), train_loss_list)
plt.title("train loss")
plt.savefig("../images/ch5-train-loss.jpg")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.plot(iter_list, train_acc_list)
plt.plot(iter_list, test_acc_list)
plt.legend(["train", "test"])
plt.title("accuracy")
plt.savefig("../images/ch5-train-test-acc.jpg")
```
